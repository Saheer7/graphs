```python

###########################################################
####### Loading the Data set with Tkinter & Pandas#########
###########################################################
# Loading the data set

# The GUI to get file name
import tkinter                     #GUI package
from tkinter.filedialog import askopenfilename    #For selection of file(dataset) in windows explorer
from tkinter.filedialog import askdirectory       #For selection of folder in windows explorer




root=tkinter.Tk()         #Creating Object for TK interface      
filename = askopenfilename(title="Please select the csv file of the data")   #Browsing & selecting the required file
print(filename)                                                              #Prints the location with file name
directory= askdirectory(title="Please mark location of folder")              #Browsing & selecting the folder location
print(directory)                                                             #Prints the location of folder to save graphs
root.destroy()                                                               #Closing/destroy object of TK interface

```

```python
# Panda library to read csv file    
df=pd.read_csv(filename)         #Reads specified filename & storing in dataframe  

df.head(5)   #Top 5 elements
```
